Import libraries

```python
import pandas as pd
import numpy as np
```

Load dataset

```python
df = pd.read_csv("/mutual_fund_data.csv")
```

Understand the dataset

```python
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.describe())
```

Clean column names

```python
df.columns = df.columns.str.strip()

df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(" ","_")
```

```python
print(df.columns)
```

Find missing values

```python
print(df.isnull().sum())
```

```python
df["average_aum_cr"] = df["average_aum_cr"].fillna(
    df["average_aum_cr"].median()
)
```

```python
print(df.isnull().sum())
```

Remove duplicates

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

Clean text values

```python
df["scheme_category"] = (
    df["scheme_category"]
    .str.strip()
    .str.lower()
)

print(df["scheme_category"])
```

Convert dates

```python
df["launch_date"] = pd.to_datetime(
    df["launch_date"],
    errors="coerce"
)
```

```python
df["launch_year"] = df["launch_date"].dt.year

df["launch_month"] = df["launch_date"].dt.month

print(df["launch_year"])
```

Create useful features using NumPy

```python
df["aum_level"] = np.where(
    df["average_aum_cr"] > 5000,
    "Large",
    "Small"
)

print(df["aum_level"])
```

Top 10 funds

```python
top10 = df.sort_values(
"average_aum_cr",
ascending=False
)

print(top10.head(10))
```

Fund launch trend

```python
print(
df["launch_year"]
.value_counts()
.sort_index()
)
```

```python
df.to_csv(
"cleaned_mutual_funds.csv",
index=False
)
```
